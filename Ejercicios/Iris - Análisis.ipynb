```python
# Gráfico de corelación
dt = df[df.columns[1:]]

corr = dt.corr()

# Quitamos el triangulo superior para facilitar la lectura
bool_upper_matrix = np.tril(np.ones(corr.shape)).astype(np.bool)
corr = corr.where(bool_upper_matrix)


# Utilizamos el heatmap de seaborn para visualizar
sns.heatmap(corr, cmap="BrBG",
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

```python
g = sns.lmplot(x="SepalLengthCm", y="SepalWidthCm", hue="Species",
               truncate=True, height=5, data=df)
g.set_axis_labels("Sepal length (cm)", "Sepal width (cm)")
```

```python
g = sns.lmplot(x="PetalLengthCm", y="PetalWidthCm", hue="Species",
               truncate=True, height=5, data=df)
g.set_axis_labels("Petal length (cm)", "Petal width (cm)")
```
