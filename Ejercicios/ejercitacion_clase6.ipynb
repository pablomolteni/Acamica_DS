### Ej 1
- cargar dataset de properati e importar librerias
- hacer un grafico de lineas con la evolucion de los precios segun las fechas .

```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
```

```python
df = pd.read_csv ('datos_properati.csv', parse_dates = ['created_on'])
df.head()
```

```python
df['Year'] = pd.DatetimeIndex(df['created_on']).year
```

```python
df.columns
df.count()
```

### Ejercicio 2

- hacer un grafico de caja con los precios en usd por **tipo de propiedad** para el año 2018.

```python
df_filtered = df[(df['Year']== 2018) & (df['price_usd_per_m2']<10000)]

df_filtered
```

```python
plt.figure(figsize=(16, 10))
ax = sns.boxplot(x='property_type', y='price_usd_per_m2', data=df_filtered)
```

### Ejercicio 3
- encontrar la cantidad de propiedades de cada barrio, que tienen un precio mayor a 650.000 o mas de 3 habitaciones
- hacer un grafico que lo demuestre

```python
# https://seaborn.pydata.org/examples/horizontal_barplot.html
```

```python
df_filtered2 = df[(df['price_aprox_usd']>650000) | (df['rooms']>3)]
df_filtered2
```

```python
plt.figure(figsize=(16,10))


ax = sns.countplot(y="barrio",
                   orient = "h",
                   order = df_filtered2['barrio'].value_counts().index,
                   data=df_filtered2)
```
